The library containing the environment is called a Gym. It is used extensively for simulation experiments in Reinforcement learning.It provides:
an interface to create RL environments
a collection of environments for use(gym-control, atari etc)

Steps associated with RL algo in gym:
Create environment : gym.make()
Reset the environment to its initial state = observation = env.reset()
Get an action using our model
Using env.step(action), perform action in the environment and this returns the new state, reward, info about episode termination.

First RL agent : taking random action from action sample space

```python
env = gym.make("LunarLander-v2")  # OPENAI gym has a collection of RL environments to tinker around.
observation = env.reset()
for _ in range(20):
  action=env.action_space.sample()
  print("Action taken: ", action)

  observation,reward,done,info = env.step(action)

  if done:
    print("Environment reset! ")
    observation = env.reset()
  
```

some details about the environment!

```python
env = gym.make("LunarLander-v2")
env.reset()

print(" ________ Observation space ____________")
print("Observation Space Shape", env.observation_space.shape)
print("Sample observation", env.observation_space.sample())
```

We see with Observation Space Shape (8,) that the observation is a vector of size 8, where each value contains different information about the lander:

Horizontal pad coordinate (x)
Vertical pad coordinate (y)
Horizontal speed (x)
Vertical speed (y)
Angle
Angular speed
If the left leg has contact point touched the land
If the right leg has contact point touched the land

```python
print("-------------Action space ------------------")
print("Action space shape ", env.action_space.n)
print("Action space sample", env.action_space.sample())
```

creation of vectorized environment helps us train agent in diverse environments together.

```python
env = make_vec_env("LunarLander-v2", n_envs=16)

```

To solve the problem a popular DL library is used called StableBaselines 3, which uses SOTA algorithm "Proximal Policy Optimization" which combines Policy based learning and Value based learning. 

since we have vectors as input, we use MlpPolicy method to train the network to learn. If we had got images as inputs, we use CNNPolicy.

```python
model = PPO(policy="MlpPolicy", env=env,n_steps=1024,batch_size=64,n_epochs=4,gamma=0.999, gae_lambda=0.98,ent_coef=0.01,verbose=1,)
```

```python
model.learn(total_timesteps=1000000)
```

```python
model_name="ppo_landerv2"
model.save(model_name)
```

```python
eval_env = gym.make("LunarLander-v2")
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=10,deterministic=True)
```
